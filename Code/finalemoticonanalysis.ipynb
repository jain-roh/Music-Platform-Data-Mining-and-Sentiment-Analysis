```python
emojis = []
meanings = []
for items in emoji_file['emoji']:
    emojis.append(items)
for items in emoji_file['meaning']:
    meanings.append(items)
```

```python
zipbObj = zip(emojis, meanings)
```

```python
dictOfEmojis = dict(zipbObj)
```

```python
for index in data_file.index:
  text = data_file.loc[index,"Review"]
  d_new = {}
  for key in dictOfEmojis:
    if key in text:
      count = text.count(key)
      meaning = dictOfEmojis[key]
      d_new[meaning] = count
  for meaning in d_new:
    count = d_new[meaning]
    data_file.loc[index, meaning] = count
```
